```python
import pandas as pd
```

```python
messages = pd.read_csv('SMSSpamCollection.txt', sep='\t', names=['label', 'message'])
```

```python
messages
```

## Data Cleaning and Preprocessing

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []

for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    
    corpus.append(review)
```

```python
corpus
```

## For Train Test Split

```python
# Dependent (Output) Features
y = pd.get_dummies(messages['label'])
```

```python
y
```

```python
y = y.iloc[:, 0].values
```

```python
y.shape
```

```python
y
```

```python
# To do Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

## To Create Bag Of Words

```python
# To create the BoWs Model
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# For Binary BoW enable binary = True
cv = CountVectorizer(max_features=2500, ngram_range=(1, 2))
```

```python
X_train
```

```python
len(X_train), len(y_train)
```

```python
# Independent Features
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
np.set_printoptions(edgeitems=30, linewidth=100000,
                    formatter=dict(float=lambda x: "%.3g" % x))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB()
```

```python
spam_detect_model
```

```python
spam_detect_model = spam_detect_model.fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

## To create the TF-IDF Model
